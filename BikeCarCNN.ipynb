```python
import tensorflow as tf
from tensorflow.keras import layers,models
import matplotlib.pyplot as plt
from collections import Counter
from keras.preprocessing.image import ImageDataGenerator
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
file_path='C:/Users/Arnav/Desktop/Arnav/BIke Car CNN/archive (1)/Car-Bike-Dataset'
```

```python
batch_size=32
img_height=256
img_width=256
```

```python
data_generator=ImageDataGenerator(rescale=1/255.,
                                  rotation_range=90,
                                  width_shift_range=0.1,
                                  horizontal_flip=True,
                                  vertical_flip=True,
                                  validation_split=0.2)
```

```python
train_set=data_generator.flow_from_directory(file_path,
                                             class_mode='binary',
                                             target_size=(img_height,img_width),
                                             shuffle=True,
                                             batch_size=batch_size,
                                             subset='training')
```

```python
test_set=data_generator.flow_from_directory(file_path,
                                            class_mode='binary',
                                            target_size=(img_height,img_width),
                                            shuffle=False,
                                            batch_size=batch_size,
                                            subset='validation')
```

```python
train_counter = Counter(train_set.classes)
test_counter = Counter(test_set.classes)
print(train_counter.items())
print(test_counter.items())
```

```python
tr_keys = list(train_counter.keys())
te_keys = list(test_counter.keys())
vals = [train_counter[k] for k in tr_keys]
vals_te = [test_counter[j] for j in te_keys]
```

```python
class_names_t = ["Bike", "Car"]
n_images = 10
plt.figure(figsize=(10, 10))
# DictionaryIterator
images, labels = train_set.next()
for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i])
    plt.title(class_names_t[labels[i].astype("uint8")])
    plt.axis("off")
    i+=1
    if i>=(n_images+1):
        break
plt.tight_layout()
plt.show()
```

```python
model = models.Sequential([
    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),
    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),
    layers.Conv2D(128, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
model.build(input_shape = (None,256,256,3))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),metrics=['accuracy'])
model.fit(train_set, validation_data=test_set, epochs=15, shuffle = True, batch_size = 32)
```

```python
eva1 = model.evaluate(test_set)

print('Test Accuracy of model: {}'.format(eva1[1]))
```
